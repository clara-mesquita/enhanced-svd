```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leia o arquivo de resultados
df = pd.read_csv("grid_search_imputation_results.csv")
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=df, x="n_lines", y="nrmse", hue="method", style="missing_rate", marker="o"
)
plt.title("NRMSE vs Janela 10\% Dados Faltantes")
plt.ylabel("NRMSE")
plt.xlabel("Janela")
plt.grid(True)
plt.show()
```

```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cria pasta plots se não existir
os.makedirs("plots", exist_ok=True)

# Localiza todos os arquivos gerados (padrão: results_datasetX_missingY.csv)
csv_files = glob.glob("results_dataset*_missing*.csv")

for csv_file in csv_files:
    # Extrai dataset_id e missing_rate do nome do arquivo
    parts = csv_file.split("_")
    dataset_id = parts[1].replace("dataset", "")
    missing_pct = parts[2].replace("missing", "").replace(".csv", "")

    df = pd.read_csv(csv_file)

    # Ajusta legenda para plot
    percent_missing = int(missing_pct)
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x="n_lines", y="nrmse", hue="method", marker="o")
    plt.title(
        f"NRMSE vs Janela\nDataset {dataset_id} - {percent_missing}% Dados Faltantes"
    )
    plt.ylabel("NRMSE")
    plt.xlabel("Tamanho da Janela (n_lines)")
    plt.grid(True)
    plt.tight_layout()
    plot_filename = (
        f"plots/nrmse_vs_janela_dataset{dataset_id}_missing{percent_missing}.png"
    )
    plt.savefig(plot_filename)
    plt.close()
    print(f"Salvo: {plot_filename}")
```
